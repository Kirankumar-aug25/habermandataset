```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sns
df = pd.read_csv(r"D:\haberman.csv")
```

```python
df.columns
```

```python
df.columns = ["age","year","nodes","status"]
```

```python
df['status']
```

```python
mp.hist(df["age"],label="Age")
mp.legend()
mp.title("Histogram plot for detcteted age.")
```

```python
mp.hist(df["year"],label="year")
mp.legend()
mp.title("Histogram plot for detcteted year.")
```

```python
mp.hist(df["nodes"],label="nodes")
mp.legend()
mp.title("Histogram plot for detcteted nodes.")
```

```python
sns.FacetGrid(df, hue="status", size=5).map(sns.distplot, "nodes").add_legend() 
    

mp.title("PDF plot for detcteted Nodes & Status.")
mp.show()


#Observation: In this plot patient survive if less nodes detected by high peck & 
  #-unsurvive if more nodes detected from dataset, you can make conclusion.
```

```python
sns.FacetGrid(df, hue="status", size=5).map(sns.distplot, "age").add_legend()
   
   
mp.title("PDF plot for detcteted age & Status.")
mp.show()
#Observation: Unpredictable anything equal number of density in each data point and PDF of status 1 
#& 2 classification overlap on each other.
```

```python
sns.FacetGrid(df, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend()
mp.title("PDF plot for detcteted Year & Status.")
mp.show()
```
